```julia
using IntervalConstraintProgramming, ValidatedNumerics
```

```julia
setrounding(Interval, :wide)
```

```julia
ring = @constraint 1 <= x^2 + y^2 <= 2
X = IntervalBox(-∞..∞, -∞..∞)
```

```julia
@time paving = setinverse(ring, X, 0.001)
```

```julia
include("draw_boxes.jl")
```

```julia
draw(paving)
```
