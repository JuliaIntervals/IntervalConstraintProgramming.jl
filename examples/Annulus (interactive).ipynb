```julia
using IntervalConstraintProgramming, ValidatedNumerics
using Interact
using Plots
gr()  # GR backend for Plots.jl
```

```julia
C = @constraint 1 <= x^2 + y^2 <= 2
X = IntervalBox(-∞..∞, -∞..∞)
```

```julia
@manipulate for n in slider(0.5:0.5:10, value=0.5)

    paving = pave(C, X, 2.0^(-n))
    plot(paving.inner, aspect_ratio=:equal, label="inside")
    plot!(paving.boundary, c="gray", alpha=0.1, ann=[(0, 0, string(Vol(paving))), (0, 0.5, "eps = 2^(-$n)")], label="boundary")
        
end
```
